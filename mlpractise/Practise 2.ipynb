```python
#Categorical variables and One Hot Encoding
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("car.csv")
df1
```

```python
dummy = pd.get_dummies(df1.model)
dummy
```

```python
merged = pd.concat([df1,dummy],axis = 'columns')
merged
```

```python
final = merged.drop(['model'],axis = 'columns')
final
```

```python
final = final.drop(['Mercedez Benz C class'], axis ='columns')
final
```

```python
x = final.drop(['Mileage'],axis = 'columns')
x
```

```python
y = final.Mileage
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.predict([[2000,6,0,0,1]])
```

```python
model.score(x,y)
```

```python
model.predict([[15000,4,1,0,1]])
```

```python
#USING SKLEARN ONE HOT ENCODER
from sklearn.preprocessing import LabelEncoder
oh = LabelEncoder()
```

```python
h = df1
h.model = oh.fit_transform(df1.model)
h
```

```python
X = h[['price','Age(yrs)']].values
X
```

```python
y = h.Mileage.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
v = ColumnTransformer([('model',OneHotEncoder(),[0])],remainder = 'passthrough')
```

```python
X = v.fit_transform(X)
X
```

```python
X = X[:,1:]
```
